# Manipulação de String e Intro (Ex. Regular)

- Expressões Regulares auxiliam nas validações


```python
# str.replace serve para tirar os sinais ">>" regex=False serve para informar que não é uma expressão regular str.strip para tirar os vazios
df['times'] = df['times'].str.replace('>>', '', regex=False).str.strip()
```

```python
df[['times', 'V']]
```

```python
import matplotlib as plt 
```

```python
grafico_percentual_vitorias = df.groupby('times')[['percentural']].sum().sort_values('percentural')
```

```python
# kind= voce escolhe o tipo de grafico deseja
grafico_percentual_vitorias.plot(kind='barh', figsize=(14,10), color='blue')
grafico_percentual_vitorias
```
